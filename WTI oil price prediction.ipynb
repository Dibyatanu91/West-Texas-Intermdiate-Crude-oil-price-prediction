# Introduction

**Background**

As one of the most important commodities in the global marker, crude oil plays an important role in economic, political and technology world. In order to have a vision for the volatility of the energy resource market, forecasting trends of crude oil prices would be of much value. This would benefit not only the government and energy departments in facilitating policy-making[1], but also  provide guidance for economic development.[2] Therefore, it will be valuable to predict crude oil price in an accurate way.

There are 2 common predicting methods that are used by most researchers. One is from the qualitative angle [3] while the other is from quantitative angle, such as econometric model and statistical model.[4, 5] And most scholars adoopted the latter.

However, non-linearity exists in crude oil market and it brings difficulty to predict the market movement. Neural network technique provides a favorable tool for nonlinear time series forecasting. 

**Purpose Statement**

In this project, I will use the deep learning knowledge of recurrent neural network (e.g. LSTM, GRU, etc.) and feed forward neural network (dense layer), to perform time series forecasting for WTI crude oil prices. And from the prediction given by the models, to provide business insights for the crude oil industry.


# Data Description

The movements of crude oil price are impacted by diverse influencing factors. The dynamic complicated interactions among these factors would result in the mysterious behavior of the crude oil price movement.[6] In the anaylsis, I am taking the factors of relevant energy resource prices and oil-sensitive stock prices into consideration, since I believe that these 2 aspects are strongly connected with the movements of crude oil prices.

I obtained data from different sources and made my own data:

### Energy Resource Price

**1) Crude Oil Prices:** For the crude oil prices, I fetched the WTI (Western Texas Intermidiate) crude oil price daily history data from an API of U.S. Energy Infomation Administration (EIA). The oil prices dataset is covering the time from 1980s to 2020.

**2) Propane & Natural Gas:** 

Besides crude oil, propane and natural gas are popular and main energy sources.

Propane is a byproduct of crude oil and natural gas processing, but the larger portion of propane produced comes from the crude oil refineries. Propane competes with other crude oil-based fuels, and the price of crude oil has a larger influence on the price of propane than that of natural gas.[7]

As for natural gas, according to a correlation study between crude oil other commodities from EIA, a correlation coefficient of 0.25 is shown between crude oil and natural gas, which indicates that a change in oil price can account for 25% of the change in natural gas prices (on average, throughout the study period).[8] This is showing that natural gas price is closely connecting with crude oil price as well.

So I also searched for propane and natural gas daily price data, as predictors to forecast crude oil price. The data of these two energy sources are also retrieved from EIA, covering the time range from 1990s to 2020.



###Oil-sensitive Stock

As stock prices are not subject to revision, they are reflecting timely market information. The stock price time series are readily available, can potentially be a valuable predictor, and thereby help to improve the accuracy of forecasts of the price of crude oil. [9] Hence for the predictors, I also included oil-sensitive stock prices. 

**3) Oil Campany Stock Prices:** The stock prices of oil companies are very important representation of how well the oil industry is performing. I included the stock prices from Yahoo Finance for the following oil companies, as predictors of the voltality of the oil market:

*   British Petroleum (BP): A British multinational oil and gas company. The world's 6th largest oil and gas company. 
*   ExxoMobil (XOM): An American multinational oil campany headquartered in Texas. One of the world's big oil companies.
*   Chevron (CVR): One of the big oil companies that are dominating the global petroleum industry. In 2018, the company produced an average of 791M barrels of net oil-equivalent per day in U.S.

In the stock prices datasets, I took the "adjclose" as "stock price" I used in predictors, since it's the closing price after adjustments for all applicable splits and dividend distributions. The stock price history is covering from 1997 to 2020.

**4) Solar Company Stock Prices:** The share prices of solar companies are closely correlated to the price of crude oil. This probably reflects the capital markets' belief that the demand of solar electricity is mainly driven by the price of oil (as a proxy for the price energy).[10] Therefore, I selected NextExtra Energy (NEE), a U.S. solar company and included its 1997-2020 stocks prices into the final dataset.

**Time Series Adjustment:** After collecting all the data sources, I adjusted the time range of the dataset and made a time series from Jan 8, 1997 to Nov 3, 2020.

```
# split a multivariate sequence into samples
def split_sequences(sequences, n_steps):
	X, y = list(), list()
	for i in range(0,len(sequences)): # UPDATED FOR N_STEPS
		# find the end of this pattern
		end_ix = i + n_steps
		# check if beyond the dataset
		if end_ix > len(sequences):
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
		X.append(seq_x)
		y.append(seq_y)
	return np.array(X), np.array(y)
```

```
# Setting time look back of 10
n_steps = 10
X_train, y_train = split_sequences(np.array(df_train), n_steps)
X_test, y_test = split_sequences(np.array(df_test), n_steps)
# check work
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape) # looks good!
```

# SimpleRNN (Model 1)
This one is base line model with just one simple RNN layer. I intend to compare other models against this to see which one is the best.

```
# now let's build a model
# NEED TO UPDATE FOR CLASSIFICATION

# define
n_steps = X_train.shape[1]
n_features = X_train.shape[2]

# define model
model = Sequential()
model.add(SimpleRNN(60, input_shape=(n_steps,n_features), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['acc'])


es = EarlyStopping(monitor='val_acc', mode='max', 
                   patience=10, 
                   verbose=1,
                   restore_best_weights=True)

# fit model (uses early stopping)
model.fit(X_train, y_train, 
          epochs=500, 
          batch_size=5, 
          validation_split=0.2, # val is a random 20% of the data since shuffle = True
          verbose=1,
          callbacks=[es],
          shuffle=True)
```

```
# make a prediction
pred = model.predict(X_test)# the pred
print(pred) # round them!

pred = np.round(pred,0)
pred
```

```
# confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(confusion_matrix(y_test, pred)) # looks pretty good!
print(classification_report(y_test, pred)) 
```

# LSTM with drop out layers (Model 3)
This is a much more complex model with couple of LSTM layer with Dropouts



```
# define what you need
n_steps = X_train.shape[1]
n_features = X_train.shape[2]

# define model
model = Sequential()
model.add((LSTM(64, return_sequences=True, activation='relu',
                input_shape=(n_steps, n_features))))
model.add(Dropout(0.1))
model.add(Bidirectional(LSTM(32,activation='relu')))
model.add(Dropout(0.1))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation='sigmoid'))
model.summary()

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# early stopping
es = EarlyStopping(monitor='val_accuracy', 
                   mode='max', 
                   patience=20, 
                   verbose=1,
                   restore_best_weights=True)

# fit model, if you run locally and remember the seed, you will be able to rerun 6 epochs and get the best val accuracy
model.fit(X_train, y_train, 
          epochs=100, 
          batch_size=5, 
          validation_split=0.2, 
          verbose=1,
          callbacks=[es],
          shuffle=True)
```

# Mix bag monstor model (Model 4)
This is model with convolutional layer followed by LSTM and GRU. This the most complex monstor of all models.

```

# since this is a univariate problem, n_features will be 1 (defined this before)

# define model
model = Sequential()
model.add(Conv1D(filters=128, kernel_size=3, input_shape=(n_steps,n_features))) # notice how input shape goes in first layer
model.add(MaxPooling1D(2))
model.add(Bidirectional(LSTM(30,
                            return_sequences=True, # remember, if stacking layers,  need to return sequences!
                            activation='relu',
                            recurrent_dropout=0.2)))
model.add(GRU(20, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['acc'])
model.summary()

es = EarlyStopping(monitor='val_acc', mode='max', 
                   patience=10, 
                   verbose=1,
                   restore_best_weights=True)

# fit model (uses early stopping)
model.fit(X_train, y_train, 
          epochs=500, 
          batch_size=5, 
          validation_split=0.2, # val is a random 20% of the data since shuffle = True
          verbose=1,
          callbacks=[es],
          shuffle=True)
```

```
# make a prediction
pred = model.predict(X_test)# the pred
print(pred) # round them!

pred = np.round(pred,0)
print(pred) # run all if  get an error...

# confusion matrix -=
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
plt.figure(figsize=(20,10))
print(confusion_matrix(y_test, pred)) =
print(classification_report(y_test, pred)) 

# show timeseries plot on the train and validation data
plt.plot(np.arange(X_test[900:].shape[0]), y_test[900:], color='blue') # actual data
plt.plot(np.arange(X_test[900:].shape[0]), pred[900:], color='grey') # predicted data
plt.suptitle('Test Results')
plt.xlabel('Time')
plt.ylabel('Increase?')
plt.show()
```
